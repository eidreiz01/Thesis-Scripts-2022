```python
for post in litecoin.hot(limit=5):
    print(post.title)
    print()
```

```python
bitcoin_posts = bitcoin.top(limit=None)
ethereum_posts = ethereum.top(limit=None)
litecoin_posts = litecoin.top(limit=None)

bitcoin_posts_dict = {"Title": []}
for post in bitcoin_posts:
    bitcoin_posts_dict["Title"].append(post.title)
bitcoin_posts = pd.DataFrame(bitcoin_posts_dict)
ethereum_posts_dict = {"Title": []}
for post in ethereum_posts:
    ethereum_posts_dict["Title"].append(post.title)
ethereum_posts = pd.DataFrame(ethereum_posts_dict)
litecoin_posts_dict = {"Title": []}
for post in litecoin_posts:
    litecoin_posts_dict["Title"].append(post.title)
litecoin_posts = pd.DataFrame(litecoin_posts_dict)
```

```python
reddit_df = pd.concat([bitcoin_posts, ethereum_posts, litecoin_posts])
reddit_df = reddit_df.sample(frac=1).reset_index(drop=True)
print(reddit_df.shape)
reddit_df.head()
```

```python
reddit_df.to_csv('reddit.csv')
```
